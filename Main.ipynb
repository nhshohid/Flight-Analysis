```python
# Iterating over 3 years of flight data to create a dataframe, 
# choose which column data to keep while also rearranging them
# and append it to a list containing all the data frames

years = ["2020","2021","2022"]
dataframes = []
cols = [
    "FlightDate",
    "Year", "Quarter", "Month", "DayofMonth", "DayOfWeek",
    "Airline",
    "Origin", "OriginCityName", "OriginState",
    "Dest", "DestCityName", "DestState",
    "Cancelled", "Diverted",
    "CRSDepTime", "DepTime", "DepDelay",
    "CRSArrTime", "ArrTime", "ArrDelay",
    "DepDel15", "ArrDel15",
    "AirTime", "CRSElapsedTime", "ActualElapsedTime",
    "Distance",
    "TaxiOut", "TaxiIn"
    ]

for year in years:
    df = pd.read_csv("Combined_Flights_" + year + ".csv")
    df = df[cols]
    dataframes.append(df)
```

```python
df_flightdata = pd.concat(dataframes)
```

```python
# grouping flights by airline, calculating the average delay time, and display the results in descending order.
df_flightdata.groupby("Airline")["DepDelay"].mean().sort_values(ascending = False).round(2)
```

JetBlue Airways had the highest average delay time of about 15 minutes and 14 seconds.

ExpressJet Airlines Inc. had the lowest average delay time of about 0.01 seconds.

```python
# displayng JetBlue Airways average delay times over the years
df_flightdata[df_flightdata["Airline"] == "JetBlue Airways"].groupby("Year")["DepDelay"].mean().round(2)
```

JetBlue Airways experienced a significant increase
in average departure delays from 2020 to 2022.

Average delays increased from approximately 1.47 minutes
in 2020 to 26.01 minutes in 2022, indicating worsening
operational performance following the COVID-19 pandemic period.

```python
# filtering out smaller airports and then checking their average delay times

airport_counts = df_flightdata["Origin"].value_counts()
large_airports = airport_counts[airport_counts > 10000].index
df_filtered = df_flightdata[df_flightdata["Origin"].isin(large_airports)]
df_filtered.groupby("Origin")["DepDelay"].mean().sort_values(ascending=False).round(2)
```

Aspen/Pitkin County Airport (ASE) had the highest average delay time
